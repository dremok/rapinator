```python
from rapinator import Genius
from rapinator import get_all_rappers
```

```python
genius = Genius('dg-YwkG0YhCgRm6ZEPr7hQDlFgyxDGIXncArkjy8bLglxpkW7dBT2aZFaijDrB6W')
```

```python
rappers = []
for rapper in get_all_rappers():
    if '[' in rapper:
        rappers.append(rapper[:rapper.index('[')])
    else:
        rappers.append(rapper)
rappers = sorted(list(set(rappers)))
```

```python
def download_artist(artist_name):
    try:
        genius.scrape_artist(artist_name, max_songs=50)
    except Exception:
        pass
```

```python
import concurrent.futures
```

```python
with concurrent.futures.ThreadPoolExecutor(max_workers=5) as executor:
    executor.map(download_artist, rappers[rappers.index('Rah Digga'):])
```

```python
import os
from pathlib import Path

data_folder = Path('/Users/max.leander/src/python/rapinator', 'data')
artist_folders = os.listdir(data_folder)
print(artist_folders)
```

```python
training_data = []

for artist_name in artist_folders:
    artist_folder = Path(data_folder, artist_name)
    if not os.path.isdir(artist_folder):
        continue
    text_files = os.listdir(artist_folder)
    for tf in text_files:
        with open(f'{artist_folder}/{tf}') as f:
            try:
                lines = [line for line in f.readlines()]
            except Exception:
                continue
            text = f'<<{artist_name} - {tf.split(".txt")[0]}>>\n' + '\n'.join(lines)
            training_data.append(text)
            print(f'Songs added to training data: {len(training_data)}')
```

```python
join_string = f"\n{'*'*50}\n"
with open(Path(data_folder, 'rapinator_training_data.txt'), 'w') as f:
    f.write(join_string.join(training_data))
```
